```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Powergrid DCF based on the Excel sheet values
actual_fcf = {
    2012: 8425.0,
    2013: 10837.0,
    2014: 12967.0,
    2015: 14927.0,
    2016: 18184.0,
    2017: 22450.0,
    2018: 25961.0,
    2019: 26920.0,
    2020: 32886.0,
    2021: 34906.0,
    2022: 35826.0,
    2023: 38580.0,
    2024: 39312.0,
    2025: 36223.0,
    2026: 40931.0
}

growth_rate = 0.065
discount_rate = 0.12
margin_of_safety = 0.20
forecast_years = list(range(2017, 2030))

cash_reserve = 89631.0
debt = 119892.0
shares_outstanding = 930.0

forecast_fcf = {2017: actual_fcf[2017]}
for year in forecast_years[1:]:
    forecast_fcf[year] = forecast_fcf[year - 1] * (1 + growth_rate)

discounted_fcf = {
    year: forecast_fcf[year] / ((1 + discount_rate) ** (year - forecast_years[0]))
    for year in forecast_years
}

terminal_value = 339279.7165
terminal_value_after_debt = 309018.7165
terminal_value_after_mos = 247214.9732
fair_price_without_mos = 332.2781898
fair_price_with_mos = 265.8225518

summary_df = pd.DataFrame({
    "Year": forecast_years,
    "Actual FCF (Cr)": [actual_fcf.get(y, np.nan) for y in forecast_years],
    "Predicted FCF (Cr)": [forecast_fcf[y] for y in forecast_years],
    "Discounted FCF (Cr)": [discounted_fcf[y] for y in forecast_years],
})

pd.set_option("display.float_format", lambda x: f"{x:,.2f}")
# print(summary_df)
# print("\n--- Powergrid DCF Summary ---")
# print(f"Growth rate: {growth_rate:.2%}")
# print(f"Discount rate: {discount_rate:.2%}")
# print(f"PV of forecast FCF: ₹{sum(discounted_fcf.values()):,.2f} Cr")
# print(f"Terminal value: ₹{terminal_value:,.2f} Cr")
# print(f"Terminal value after debt: ₹{terminal_value_after_debt:,.2f} Cr")
# print(f"Terminal value after margin of safety: ₹{terminal_value_after_mos:,.2f} Cr")
# print(f"Cash reserve: ₹{cash_reserve:,.2f} Cr")
# print(f"Debt: ₹{debt:,.2f} Cr")
# print(f"Shares outstanding: {shares_outstanding:,.2f} Cr")
# print(f"Fair price without margin of safety: ₹{fair_price_without_mos:,.2f}")
# print(f"Fair price with margin of safety: ₹{fair_price_with_mos:,.2f}")

#######################

from IPython.display import display, Markdown

display(summary_df)

display(Markdown(f"""
## PowerGrid DCF Summary

| Metric | Value |
|---|---:|
| Growth rate | **{growth_rate:.2%}** |
| Discount rate | **{discount_rate:.2%}** |
| PV of forecast FCF | **₹{sum(discounted_fcf.values()):,.2f} Cr** |
| Terminal value | **₹{terminal_value:,.2f} Cr** |
| Terminal value after debt | **₹{terminal_value_after_debt:,.2f} Cr** |
| Terminal value after margin of safety | **₹{terminal_value_after_mos:,.2f} Cr** |
| Cash reserve | **₹{cash_reserve:,.2f} Cr** |
| Debt | **₹{debt:,.2f} Cr** |
| Shares outstanding | **{shares_outstanding:,.2f} Cr** |
| Fair price without margin of safety | **₹{fair_price_without_mos:,.2f}** |
| Fair price with margin of safety | **₹{fair_price_with_mos:,.2f}** |
"""))

##################################

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=list(actual_fcf.keys()),
    y=list(actual_fcf.values()),
    mode="lines+markers",
    name="Actual FCF"
))
fig.add_trace(go.Scatter(
    x=forecast_years,
    y=[forecast_fcf[y] for y in forecast_years],
    mode="lines+markers",
    name="Predicted FCF"
))
fig.update_layout(
    title="Powergrid: Actual vs Predicted FCF",
    xaxis_title="Year",
    yaxis_title="FCF / Operating Profit (₹ Cr)",
    template="plotly_white",
    hovermode="x unified"
)
fig.show()
```

$$ PV of FCF + Cash \; reserve - Debt
= Equity \; value - Margin\; of \; safety
= Final \; value $$

```python
pv_forecast_fcf = sum(discounted_fcf.values())

cash_reserve = 89631.0
debt = 119892.0
shares_outstanding = 930.0
margin_of_safety = 0.20

terminal_value_including_assets_and_reserves = 309018.7165
equity_value = terminal_value_including_assets_and_reserves
equity_value_after_mos = 247214.9732

fair_price_without_mos = 332.2781898
fair_price_with_mos = 265.8225518

fig = go.Figure(go.Waterfall(
    name="DCF Bridge",
    orientation="v",
    measure=["relative", "relative", "relative", "total", "relative", "total"],
    x=["PV of Forecast FCF", "Cash Reserve", "Debt", "Equity Value", "Margin of Safety", "Final Value"],
    y=[
        pv_forecast_fcf,
        cash_reserve,
        -debt,
        0,
        -equity_value * margin_of_safety,
        0,
    ],
    text=[
        f"₹{pv_forecast_fcf:,.0f} Cr",
        f"₹{cash_reserve:,.0f} Cr",
        f"-₹{debt:,.0f} Cr",
        f"₹{equity_value:,.0f} Cr",
        f"-₹{equity_value * margin_of_safety:,.0f} Cr",
        f"₹{equity_value_after_mos:,.0f} Cr",
    ],
    textposition="outside"
))

fig.update_layout(
    title=f"Powergrid DCF Valuation Bridge | Fair Value: ₹{fair_price_with_mos:,.2f} with 20% MoS",
    yaxis_title="₹ Cr",
    template="plotly_white",
)
fig.show()
```

Growth rate: 3% to 9%
Discount rate: 12% to 18%

```python
growth_rates = np.arange(0.03, 0.095, 0.005)
discount_rates = np.arange(0.12, 0.185, 0.005)

sensitivity = []

for dr in discount_rates:
    row = []
    for gr in growth_rates:
        temp_predicted_fcf = {}
        temp_discounted_fcf = {}

        for year in forecast_years:
            if year == forecast_years[0]:
                temp_predicted_fcf[year] = actual_fcf[forecast_years[0]]
            else:
                temp_predicted_fcf[year] = temp_predicted_fcf[year - 1] * (1 + gr)

            discount_period = year - forecast_years[0]
            temp_discounted_fcf[year] = temp_predicted_fcf[year] / ((1 + dr) ** discount_period)

        temp_pv_fcf = sum(temp_discounted_fcf.values())
        temp_equity_value = temp_pv_fcf + cash_reserve - debt
        temp_fair_price = temp_equity_value / shares_outstanding
        row.append(temp_fair_price)

    sensitivity.append(row)

fig = go.Figure(data=go.Heatmap(
    z=sensitivity,
    x=[f"{g:.1%}" for g in growth_rates],
    y=[f"{d:.1%}" for d in discount_rates],
    text=[[f"₹{v:.0f}" for v in row] for row in sensitivity],
    texttemplate="%{text}",
    hovertemplate="Growth: %{x}<br>Discount Rate: %{y}<br>Fair Value: %{text}<extra></extra>",
))

fig.update_layout(
    title="Powergrid DCF Sensitivity: Fair Value Without Margin of Safety",
    xaxis_title="FCF Growth Rate",
    yaxis_title="Discount Rate",
    template="plotly_white",
)

fig.show()
```
